```python
cities = pl.read_csv("./data/cities.csv", has_header=False, new_columns=["city"]).drop_nulls()
```

```python
cities.head()
```

```python
countries_to_remove = ["England", "Scotland", "Wales", "Northern Ireland"]

print(cities.height)
cities = cities.filter(~pl.col("city").is_in(countries_to_remove))
print(cities.height)
```

```python
cities = cities.with_columns(city=pl.col("city").str.replace("*", "", literal=True))
```

```python
cities = cities.with_columns(city=pl.col("city").str.to_uppercase())
cities.head()
```

```python
# city_pattern = (
#     cities.with_columns(
#         pl.col("city").str.escape_regex()
#     )
#     .get_column("city")
#     .reverse()
#     .str.join("|")
#     .item()
# )

# customers = customers.with_columns(
#     pl.col("address_clean")
#     .str.extract(rf"\n({city_pattern}),", group_index=1)
#     .fill_null("OTHER")
#     .alias("city")
# )

customers = customers.with_columns(city=pl.lit(None, dtype=pl.String))
for city in cities["city"]:
    customers = customers.with_columns(
        pl.when(
            pl.col("address_clean").str.contains(f"\n{city},")
        )
        .then(pl.lit(city))
        .otherwise(pl.col("city"))
        .alias("city")
    )

customers = customers.with_columns(
    pl.col("city").fill_null("OTHER")
)
```

```python
customers.head()
```

```python
customers.get_column("city").value_counts().sort("count", descending=True).head(20)
```

```python
sample_other = customers.filter(pl.col("city") == "OTHER").sample(5, seed=42)
for address in sample_other["address_clean"]:
    print(address, "\n")
```

```python
print(f"{customers["city"].n_unique()} cities in customer data (including OTHER)")
print(f"{cities.height} cities in city list")
```

```python
set(cities["city"]) - set(customers["city"])
```

```python
customers.filter(pl.col("address_clean").str.contains("\nHULL,"))
```

```python
customers = customers.with_columns(
    pl.when(pl.col("address_clean").str.contains("\nHULL,"))
    .then(pl.lit("HULL"))
    .otherwise(pl.col("city"))
    .alias("city")
)
```

```python
top_20_spend = (
    customers.group_by("city")
    .agg(total_spend=pl.col("total_spend").sum())
    .sort("total_spend", descending=True)
    .head(20)
    .sort("total_spend")
)

ggplot(top_20_spend, aes(x="total_spend", y="city")) + \
    geom_bar(stat="identity") + \
    scale_x_continuous(format="£{.4~s}") + \
    labs(
        title="Total customer spend by city",
        x="Total spend",
        y="City",
    )
        
```

```python
print("Total spend for all customers:")
print(customers["total_spend"].sum())
```

```python
print("Total spend for London customers:")
print(customers.filter(pl.col("city") == "LONDON").get_column("total_spend").sum())
```

```python
print("Total spend outside London:")
print(customers.filter(pl.col("city") != "LONDON").get_column("total_spend").sum())
print("Total spend outside London (excluding OTHER):")
print(customers.filter(~pl.col("city").is_in(["LONDON", "OTHER"])).get_column("total_spend").sum())
```
